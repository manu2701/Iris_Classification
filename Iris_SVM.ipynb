```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target']=iris.target
df.head()
```

```python
iris.target_names
```

```python
df[df.target==1].head()
```

```python
df[df.target==0].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name']=df.target.apply(lambda x:iris.target_names[x])
df.head()
```

```python
df0=df[df.target==0]
df1=df[df.target==1]
df2=df[df.target==2]
```

```python
plt.xlabel('sepal length(cm)')
plt.ylabel('sepal width(cm)')
plt.scatter(df0['sepal length (cm)'],df0['sepal width (cm)'],marker='+',color='green')
plt.scatter(df1['sepal length (cm)'],df1['sepal width (cm)'],marker='+',color='blue')
```

```python
plt.xlabel('petal length(cm)')
plt.ylabel('petal width(cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'],marker='+',color='green')
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],marker='+',color='blue')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['target','flower_name'],axis='columns')
X.head()
```

```python
y=df.target
y.head()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
model=SVC()    #you can also provide values for gamma,C and also kernel
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
model.score(X_test,y_test)
```
